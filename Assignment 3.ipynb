```python
# Assignment 3

import random
import matplotlib.pyplot as plt
import time



def hatter(number): #Models the game
    hat = list(range(1,number+1))
    for i in range(number-1):

        posx = random.randint(0,len(hat)-1)
        posy = random.randint(0,len(hat)-1)


        while posx == posy:
            posy = random.randint(0,len(hat)-1)
    
        x = hat[posx]
        y = hat[posy]
    
        dif  = abs(x-y)
    
        posx, posy = sorted((posx, posy), reverse=True)
    
        hat.pop(posx)
        hat.pop(posy)
        hat.append(dif)
    return (hat)
#start = time.time()

def run(num,itera): #Runs the game itera times for different N
    results = []
    total = 0

    for i in range(itera):
    
        x = hatter(num)
    
        results.append(x)
    
   
    for i in results:
        total += int(i[0])
    #print(results)
        
    #print(total/len(results))
    return total/len(results)


print('N=3 \n')
run(3,500)
print('\n N=4 \n')
run(4,500)
print('\n N=5 \n')
run(5,500)
print('\n N=100 \n')
run(100,500)
print('\n N=2024 \n')
run(2024,500)

#end = time.time()
#print(end - start)
```

```python
def hist(): #Plots the histogram for the N=2024 case
    results = []

    for i in range(500):
    
        x = hatter(2024)
    
        results.append(x)
    
    plt.hist(results,bins=5,density=True)
    plt.xlabel('Final number')
    plt.ylabel('Frequency')
    plt.show()
    
hist()
```

```python
# Computing the expected difference after one iteration for N=2024 
sum =0 
for i in range(1,2025):
    for j in range(1,2025):
        sum += abs(i-j)
        
expectation = (1/2024**2)*sum
print(expectation)

    
```

```python
def cor(N,it): #Plots correlation between the expected final number and N
    av = []

    for i in range(1,N):
        x = run(i,it)
        av.append(x)



    plt.plot(list(range(1,N)),av)
    plt.show()
    
cor(100,50)
cor(100,400)
cor(500,20)
```
